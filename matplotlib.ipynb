```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0, 10)
y = np.arange(11, 21)
```

```python
plt.scatter(x, y, c='g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D')
plt.savefig('test.png')
```

```python
plt.plot(x, y, 'g', linestyle='dashed', linewidth=2, markersize=12)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'g')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'r')
plt.subplot(2, 2, 3)
plt.plot(x, y, 'b*-')
plt.subplot(2, 2, 4)
plt.plot(x, y, 'y--')
```

```python
x = np.arange(0, 10)
y = 2 * x + 5
plt.plot(x, y, 'g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Linear operations on x and y')
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y, 'g')
plt.title('Sine Wave form')
```

```python
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('sine')

plt.subplot(2, 1, 2)
plt.plot(x, y_cos, 'g--')
plt.title('cosine')
```

```python
x = [1, 5, 10]
y = [2, 6, 12]

x2 = [4, 6, 8]
y2 = [5, 7, 10]

plt.bar(x, y)
plt.bar(x2, y2, color='g')

plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
a = np.array([1, 1, 2, 2, 4, 3, 4, 5])
plt.hist(a)
plt.title('Histogram')
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
labels = 'Python', 'C++', 'C', 'Ruby'
colors = ['red', 'blue', 'yellow', 'green']
sizes = [23, 34, 56, 78]
explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct = "%1.1f%%", shadow=True)
plt.axis('equal')
```
